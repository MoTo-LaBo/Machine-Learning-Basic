## 応用

- 凡例表示
- legend() 関数で指定
- loc で位置を指定。「 best 」 を指定すると、一番良い場所に自動で表示

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 9, 2, 8])
plt.plot(x, y, label='label')
plt.title('Line graph(MATLAB-style)')
plt.legend(loc='best')
plt.show()
```

### 凡例位置指定
- 上 ・ 下　:　upper, lower
- 左 ・ 中央 ・ 右　:　left, center, right

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 5, 9, 2, 8])
y2 = np.array([2, 6, 7, 9, 5])
plt.plot(x, y1, label='label1')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.plot(x, y2, label='label2')
plt.title('Line graph(MATLAB-style)')
plt.legend(loc='lower right')
plt.show()
```

### sin, cos
- サイン, コサインカーブ
- plot関数で細かい点を結びつけて曲線を描画

```python
x = np.arange(0.0, 20.0, 0.1)  # 0.0 ~ 20.0 まで 0.1ずつ
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('sin,cos')
plt.legend(loc='upper center')
plt.show()
```

## 散布図
- 座標点の描画
- scatter関数を利用

- `機械学習では一番使用頻度が高い`

```python
np.random.seed(1)
x = np.random.random(50)
y = np.random.random(50)
plt.scatter(x, y)
plt.show()
```

### marker
- 散布図の点の形を指定
- marker パラメーターで指定

```python
np.random.seed(1)
x = np.random.random(50)
y = np.random.random(50)
# 最初の10個は下向き三角
plt.scatter(x[0:10], y[0:10], marker='v', label='triangle down')

# 次の10個は上向き三角
plt.scatter(x[10:20], y[10:20], marker='^', label='triangle up')

# 次の10個は正方形
plt.scatter(x[20:30], y[20:30], marker='s', label='square')

# 次の10個は星型
plt.scatter(x[30:40], y[30:40], marker='*', label='star')

# 次の10個は x型
plt.scatter(x[40:50], y[40:50], marker='x', label='x')
plt.legend(loc='best')
plt.show()
```

## 等高線
- 格子点の描画
- x,y 座標の全ての組み合わせを numpy array の meshgrid で作成

```python
x = np.arange(3) # array([0, 1, 2])
y = np.arange(5) # array([0, 1, 2, 3, 4])

# 格子点作成
xx, yy = np.meshgrid(x, y)
plt.scatter(xx, yy)
plt.show()
```

```python
print(f'xx:\n{xx}\nyy:\n{yy}')
```

### 等高線
- contour( )関数を利用
- meshgrid　の点と、高さを引数としてして
- cmap　で色の組み合わせの paramerter を指定できる

```python
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)  # -10.0 ~ 10.0 の間を　 0.2刻みの格子点
xx, yy = np.meshgrid(x, y)

# z の値を計算 : 3次元曲線ができる
z = xx**2 - yy**2

# 等高線表示
plt.contour(xx, yy, z, cmap="Blues")
plt.show()
```

### countourf( )
- 色が塗られる

```python
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)  # -10.0 ~ 10.0 の間を　 0.2刻みの格子点
xx, yy = np.meshgrid(x, y)

# z の値を計算 : 3次元曲線ができる
z = xx**2 - yy**2

# 等高線表示
plt.contourf(xx, yy, z, cmap="Blues")
plt.show()
```

## 描画 object
- 描画 object を用いた描画方法
    - 複数のグラフを表示
- subplots( )関数で object を生成
- subplots (ax), と描画 object (fig) を出力
- ax の plot method を用いて描画

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 9, 2, 8])
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Line graph(OOp-style)')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 5, 9, 2, 8])
y2 = np.array([2, 6, 7, 9, 5])
# subplots(描画数指定)
fig, axes = plt.subplots(2, figsize=(6, 6))
axes[0].plot(x, y1)
axes[1].bar(x, y2)
axes[0].set_title('bar graph')
axes[1].set_title('line graph')
plt.show()
```

## 複数のグラフの描画
- subplots( ) にncolsで横方向に複数のグラフを描画可能 

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 5, 9, 2, 8])
y2 = np.array([2, 6, 7, 9, 5])
# subplots(横方向)
fig, axes = plt.subplots(ncols=2)
axes[0].plot(x, y1)
axes[1].barh(x, y2)
axes[0].set_title('line graph')
axes[1].set_title('bar graph')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 5, 9, 2, 8])
y2 = np.array([2, 6, 7, 9, 5])
# subplots(描画数指定, size)
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y1)
axes[1].barh(x, y2)
axes[0].set_title('line graph')
axes[1].set_title('bar graph')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
x1 = np.random.random(50)
y1 = np.array([3, 5, 9, 2, 8])
y2 = np.array([2, 6, 7, 9, 5])
y3 = np.array([3, 5, 9, 2, 8])
y4 = np.random.random(50)
# subplots(描画数指定)
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes[0, 0].plot(x, y1)
axes[0, 1].bar(x, y2)
axes[1, 0].barh(x, y3)
axes[1, 1].scatter(x1, y4)
axes[0, 0].set_title('line graph')
axes[0, 1].set_title('bar graph')
axes[1, 0].set_title('barh graph')
axes[1, 1].set_title('scatter graph')
plt.show()
```

## ３次元の等高線
- 3次元のグラフの描画
- fig.gca(projection='3d')で実現可能
- contour3D
    - 3次元の等高線
- levels で分割するレベル数を指定

```python
dalta = 0.2
x = y = np.arange(-10.0, 10.0, delta)
xx, yy = np.meshgrid(x, y)
z = xx**2 - yy**2

# 描画サイズ (単位はインチ)
fig = plt.figure(figsize=(6, 6))

# 三次元グラフ描画 subplot を生成
ax = fig.gca(projection='3d')

# 等高線の表示
ax.contour3D(xx, yy, z, levels=40, cmap='Blues')

# x, y, z の描画範囲の設定
ax.set_xlim(xx.min()-1, xx.max()+1)     # x の領域より少しマージを持たせている
ax.set_ylim(yy.min()-1, yy.max()+1)     # y の領域よりも少しマージを持たせている
ax.set_zlim(-abs(z).max(), abs(z).max())  # z の絶対値の最大値としその反対を最小値とする

# 画像の保存
plt.savefig('contour3D.jpg', dpi=120)   # 画像保存
plt.show()
```

```python
delta = 0.2
x = y = np.arange(-10.0, 10.0, delta)  # -10.0 ~ 10.0 の間を　 0.2刻みの格子点
xx, yy = np.meshgrid(x, y)

# z の値を計算 : 3次元曲線ができる
z = xx**2 - yy**2

# 等高線表示
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].contour(xx, yy, z, cmap="Blues")
axes[1].contourf(xx, yy, z, cmap="Blues")
axes[0].set_title('contour line')
axes[1].set_title('contour line color')
plt.show()
```

## 等高線　label
- clabel method で指定可能

```python
# 描画size ( default = 横 8, 縦 6 )
fig = plt.figure()

# subplot の取得
ax = fig.gca()

# 等高線 label
cset = ax.contour(xx, yy, z, levels=10, cmap='Blues')

# x, y　の描画範囲
ax.set_xlim(xx.min(), xx.max())
ax.set_ylim(yy.min(), yy.max())
ax.clabel(cset, fontsize=9, inline=1)

# 画像保存
plt.savefig('contour3d_2d.jpg', dpi=120)
plt.show()
```
