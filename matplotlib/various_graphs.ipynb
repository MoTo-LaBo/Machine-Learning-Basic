# 各種のグラフ(2)

## 円グラフ

```python
x = np.array([1, 3, 5, 7, 9])
g = plt.pie(x)
```

## label を与える

```python
lbl = ['1st', '2nd', '3rd', '4th', '5th']
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl)
```

## 開始角度と回転方向
- 右向き半径を基準にして「反時計回り」　->　default の場合
    - matplotlib は「極座標系」の考え方を取り入れているから
    - 科学技術系の作図 library なの、そういう分野の常識を元にして作られているから
    -　「極座標系」(回転式座標系)は右向き半径が開始位置になっている。角度が０度
    - 角度が上がっていくと反時計回りに上がっていく特徴を持つ
    
- 理工学系の分野以外だとネガティブな影響を与える

### 開始角度・回転方向の設定
- startangle　:　開始角度の設定
- counterclock　:　時計回りにする

```python
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl, startangle=90, counterclock=False)
```

### 円弧の色の指定
- colors　:　円弧の色の指定

```python
clr = ['green', 'blue', 'cyan', 'magenta', 'yellow']
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl, colors=clr, startangle=90, counterclock=False)
```

### グレースケールにする
- 色の要素に 0 ~ 1.0 の値を文字列で与える

```python
clr = ['0.4', '0.5', '0.6', '0.7', '0.8']
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl, colors=clr, startangle=90, counterclock=False)
```

## まとめ
- <u>円グラフ</u>
    - plt.pie(配列, labels=ラベル列, 各種option)


- <u>各種option</u>
    - startangle=開始角度
    - counterclock=　True/False　:　False　を与えると時計回り
        - default が反時計回り
        
    - colors=色のデータ列
        - 色名か色コードのデータ列
        - '数値'　の並びでグレースケール　※ str型(string)

# 各種グラフ(3)

## 円グラフの続き
- 円弧のパーセンテージの表示
- 円弧の切り離し

### %(パーセンテージ)の表示
- autopct='%.1f%%'
    - pct ->  percentageの略
    - %表示桁数.小数点以下の桁数f%%
        - 小数点以下は、長くても２桁が良いとされている
        - .1f ->　１桁
        - 「表示桁数」を省くこともできる ->　その場合は自動調整
    - %表示桁数d%%
        - 整数でパーセント表示される
        - パーセント(%)がいらない場合は、末尾の　%%　をとる

```python
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl,
            autopct='%.1f%%',
            colors=clr,
            startangle=90,
            counterclock=False)
```

### 特定の円弧を切り離す
- 飛び出し量は　0 ~ 1.0 の値で指定する
    - 1.0 は半径の大きさ

```python
ex = [0, 0, 0.1, 0, 0.2]
plt.figure(figsize=(5, 5))
g = plt.pie(x, labels=lbl,      # ラベル
            autopct='%.1f%%',   # パーセンテージ
            explode=ex,         # 各円弧の飛び出し量
            colors=clr,         # カラー
            startangle=90,      # スタート位置
            counterclock=False) # 時計回り
```

## まとめ
- <u>パーセンテージの表示</u>
    - autopct='%.1f%%'　:　小数点表示の場合
    - autopct='%d%%'　　:　整数表示の場合


- <u>円弧の切り離し</u>
    - explode=距離のデータ列
    - 「距離」は半径の長さを 1 とする比率で与える

# 複数のグラフの作図

## 複数のグラフの作図

```python
x = np.linspace(-2, 2, 100)
y1 = x**2                  # 放物線
y2 = x*(x-1)*(x+1)         # 三次関数
```

### １つの平面に重ねてプロット
- plt.plot()　:　プロット関数を２回記述することにより、２つのグラフが重なってプロットされる

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

### 個別にplotする
- python の分割代入でよく用いられる
- subplots(１行, 2列)　:　１行に横並びでグラフがplotされる

```python
(fig, ax) = plt.subplots(1, 2)
```

- ( )　は省略できる

- fig　->　１つのグラフ用紙

- ax　->　個別のグラフ

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
```

- 機械学習の時に複数のdata列の相互の関係をタイル状にグラフに並べて比較する場合に重宝する

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('2 graphs')
ax[0].plot(x, y1)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('parabola')
ax[1].plot(x, y2)
ax[1].set_xlabel('x')
ax[1].set_ylabel('x*(x-1)*(x+1)')
ax[1].set_title('cubic')
```

- subplots　の場合は違う　method　が使用されるので注意

- ax[index].set_< method >(' ')

## グラフを縦方向に並べる
- 2行1列に引数を変更
- figsize も変更

```python
fig, ax = plt.subplots(2, 1, figsize=(4, 10))
fig.suptitle('2 graphs')
ax[0].plot(x, y1)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('parabola')
ax[1].plot(x, y2)
ax[1].set_xlabel('x')
ax[1].set_ylabel('x*(x-1)*(x+1)')
ax[1].set_title('cubic')
```

## 複数のグラフを配置する「行」と「列」
- 二次元的配列やグラフの配置をする事が多い
- 表の様な配置　->　行と列の概念で出来ている
- <u>基本的だがとても重要な事</u>


- fig, ax = plt.subplots(1, 2)　->　「１行２列」

## 複数の行・列に渡るplot

```python
x2 = np.linspace(-7, 7, 100)
y3 = np.sin(x2)
y4 = np.cos(x2)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,5))
ax[0, 0].plot(x2, y3)
ax[0, 0].grid()
ax[1, 0].plot(x2, y4)
ax[1, 0].grid()
ax[0, 1].plot(x, y1)
ax[0, 1].grid()
ax[1, 1].plot(x, y2)
ax[1, 1].grid()
```

## まとめ
- <u>行数と列数を指定して複数のグラフの並びを作成する</u>
    - (fig,ax)=plt.subplots(行数,列数)
    - fig　:　描画領域全体
    - ax　:　個々のグラフ,　スライス [index　] をつけて選択する


- <u>独特の方法でtitle, label を与える</u>
    - fig.suptitle('全体のタイトル')　:　全体のタイトル
    - ax[index].set_< method >(' ')
        - title　:　個別のタイトル
        - xlabel　:　横軸ラベル
        - ylabel　:　縦軸ラベル
        
        
- <u>複数の行,列に渡るグラフを作成する際は、ax　は 2次元の配列となる</u>
    - ax[2, 3].plot(x, y)
