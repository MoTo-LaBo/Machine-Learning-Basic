# 基本統計量の概念

## 基本統計量概念


#### `標本数`
- 標本 data の個数


#### `合計`
- data の値の合計値


#### `平均`
- 算術平均や相加平均とも呼ばれる
- data の合計を data の数で割ったもの
    
$$
x_1,　x_2,　... ,　x_N
$$


$$
\bar{x}　=　\frac{1}{N} ( x_1　+　x_2　+　...　+　x_N )　=　\frac{1}{N}\sum_{n=1}^N x_n
$$


#### `最小値`
- data の中の最小値


#### `中央値`
- data の個数を最大値から数えて丁度半分のところの値
- data の個数が奇数個ならば丁度半分のところの値
    - 偶数個ならば前後に数を足して丁度２で割った値
        - <u>外れ値の影響を受けにくい</u>
        - `中央値 : 4`
        $$1, 3, 3, 4, 5, 6, 7　->　4$$
            - `中央値 : 4.5`
        $$1, 2, 3, 4, 5, 7, 8, 9　->　(4+5)/2=4.5$$


#### `分散`
- data のばらつき度合いを表す
- 平均の近くにバラついている時はバラつきが少ない
- 平均から遠くに離れ他ところまでバラついている時はバラつきが大きい
- 値が大きほどバラつきは大きい
    - ($\sigma^2$　:　分散,　$n$　:　data総数,　$x_i$　:　個々の数値$(x_i　->　x_n)$　$\bar{x}$　xの平均値)


$$\sigma^2　=　\frac{1}{n}\sum_{i=1}^{n}(x_i-\bar{x})^2$$


- **母分散**
$$
\sigma^2　=　\frac{1}{N}\sum_{n=1}^N( x_n　-　\bar{x} )^2
$$


- **標本分散**
$$
\sigma^2　=　\frac{1}{N-1}\sum_{n=1}^N( x_n　-　\bar{x} )^2
$$

      

#### `標準偏差`
- **分散に　=　$\sqrt{\sigma^2}$**
    - <u>data のバラつきをを定量評価する</u>
    - *もとのスケールに戻す*
    - 分散の平方根を取ったもの
    - 分散同様に data のバラつきを表す
    - 分散よりも data の**実態に忠実**


$$
\sigma　=　\sqrt{\sigma^2}
$$


- ($\sigma$　:　標準偏差,　$n$　:　data総数,　$x_i$　:　個々の数値$(x_i　->　x_n)$　$\bar{x}$　xの平均値)
    - 中身は　**分散**
    - <u>標準偏差を見た方が実態のバラつき具合がわかる</u>


$$
\sigma　=　\sqrt{\frac{1}{n}\sum_{i=1}^{n}(x_i-\bar{x})^2}
$$


- **四分位数(しぶんいすう)**
    - data の相対的位置を見る場合に用いる
        - data を小さい順に左から並べる
        
        
        - 最小値
            - 第一四分位   （25%）
            - 第二四分位   (50%) = 中央値
            - 第三四分位点 (75%)
            
            
        - data が偶数の場合
            - data を前半・後半に分ける
            
            
        - data が奇数の場合
            - 中央値をはさんで、前半・後半に分ける
            
            
    - 第一四分位点　:　前半の data の中央値
    - 第三四分位点　:　後半の data の中央値
    - 四分位範囲　:　第三四分位数 - 第一四分位数
        - <u>役割は標準偏差と同じ</u>
            - *data量が少ない時に使用する*
            - **機械学習をする場合にはあまり四分位数は出てこない**

# 正規分布　標準偏差　$3\sigma$法

- `確率分布`
    - 確率変数がとる値とその値を取る確率の対応の様子

|サイコロの目|1|2|3|4|5|6|
|--|--|--|--|--|--|--|
|確率|$$\frac{1}{6}$$|$$\frac{1}{6}$$|$$\frac{1}{6}$$|$$\frac{1}{6}$$|$$\frac{1}{6}$$|$$\frac{1}{6}$$|


- `正規分布`
    - 平均値の発生頻度が最も高い確率分布
    - 平均値を中心に左右対称で発生頻度が高くなっている
    - ヒストグラム全体の形状は釣鐘状になる



- **正規分布を使用する理由は?**
    - 数式として扱い易い
    - 物理現象に近い値が出てくる
    - 2つの性質を兼ね備えている



- `標準偏差　:`　**$\sigma$**



- `正規分布`
    - 平均値の発生頻度が最も高い確率分布
    - 平均値を中心に左右対称で発生頻度が高くなっている
    - ヒストグラム全体の形状は釣鐘状になる



- *正規分布にする data 例*
    - 大規模な模試の点数分布
    - 全国の高校生の男女別の身長
    - サイコロを何回も投げた時の出目の合計



- <u>正規分布の場合、平均値　±　標準偏差中に観測 data が含まれる確率は **68.3%**</u>
    - **±　$2\sigma$　, ±　$3\sigma$　になると確率は上がる**

### `68 - 95 - 99.7rule`
- $\sigma$　=　標準偏差
- $\mu$　=　平均


- **正規分布では**
    - 平均から　**± １**　標準偏差には約　**68%**　のデータが　$$\mu-\sigma,　\mu+\sigma$$
    - 平均から　**± 2**　標準偏差には約　**95%**　のデータが　$$\mu-2\sigma,　\mu+2\sigma$$
    - 平均から　**± 3**　標準偏差には約　**99%**　のデータが　$$\mu-3\sigma,　\mu+3\sigma$$
        - 含まれるという統計的に分かっている。経験則がある


- <u>あくまで上記は　**正規分布**　の時だけ</u>

## $3\sigma$法
- **正規分布は何に使用できるのか？**
    - <u>一番多く出てくる頻度・回数に正規分布は出てくる</u>


- **正規分布に従っていると仮定すると...**
    - **± 3　$\sigma$**　の間に　**99.7%**　のデータが　含まれるという統計的に分かっている(経験則がある)


$$\mu-3\sigma,　\mu+3\sigma$$


#### **$3\sigma$ を閾値として設ける**
- <u>簡単に外れ値を算出・抽出する事ができる</u>


- $3\sigma$法　=　**99.7%**　のdata を取得
    - **製造業の異常検知などでも使用される**


- *Hampel 判別法というのもある*
    - Hampel Identifier
        - *$3\sigma$法を中央値 base にしたモノ*
        - *$3\sigma$法だと大きな値に引っ張られてしまう事がある*

## スケーリング

- **スケールの統一**
    - data　のバラつき　(　paramertor　:　$\omega$　(重み)　によってscaleがまちまち　)
        - data のバラつきをバラつきで割る
        - **data のバラつき　=　標準偏差**

#### `Standard Scaling`
- <u>data の標準化</u>
    - **平均　:　0**　->　平均で引くので平均が 0 になる
    - **標準偏差　:　1**　->　標準偏差を標準偏差で割るので 1 になる


$$
x_s　=　\frac{x - \bar{x}}{\sigma}
$$


- **標準化により data の scale を統一する**
    - ※ あくまで標準偏差が　１　で最大値が　１　という事ではない！
    - 1　以上になる事もある
