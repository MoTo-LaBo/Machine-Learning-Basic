### スカラー, ベクトル, 行列
- `スカラー`
    - **1行・1列**
$$x,　y,　z,　M,　N$$


- `ベクトル`
    - *複数の変数・定数・値を集めたモノ*
        - <u>スカラーが複数個集まればベクトル</u>
        - **3行・1列**
        

$$x= \left(
    \begin{array}{c}
      x_1 \\
      x_2 \\
      x_3 \\
    \end{array}
  \right)
$$


$$x= \left(
    \begin{array}{c}
      y_1 \\
      y_2 \\
      .   \\
      .   \\
      y_n \\
    \end{array}
  \right)
$$


- *T は転置 : 横向き*
$$
x^T=(1　2　3)
$$


- `行列`
    - *ベクトルに対してさらに横方向に持ったもの*
        - **3行・2列**

 
$$X=\left(
    \begin{array}{cc}
      x_{11} & x_{12} \\
      x_{21} & x_{22} \\
      x_{31} & x_{32} \\
    \end{array}
  \right)
$$


※ M, N, O, P, L 以外の大文字は行列とする


### 文字の認識
- スカラー　:　小文字の細文字
    - *変数*


-　スカラー　:　大文字の細文字
    - *定数*
    

- ベクトル　:　小文字の太文字
    - **ベクトル**
    

- 行列　:　大文字の太文字
    - **行列**

### <u>file</u>

## `単位行列`
- *斜めの方向だけ 1 それ以外を 0 の正方形の行列*


$$I=\left(
    \begin{array}{cc}
      {1} & {0} & \ldots & {0} \\
      {0} & {1} & \ldots & {0} \\
      \vdots & \vdots & \ddots & \vdots \\
      {0} & {0} & \ldots & {1} \\
    \end{array}
\right)
$$


$$I_{2*2}=\left(
    \begin{array}{cc}
      {1} & {0} \\
      {0} & {1} \\
    \end{array}
\right)
$$


$$I_{3*3}=\left(
    \begin{array}{cc}
      {1} & {0} & {0} \\
      {0} & {1} & {0} \\
      {0} & {0} & {1} \\
    \end{array}
\right)
$$


### <u>性質</u>
- **スカラーの　「　1　」 に対応**
$$
2*1　=　2
$$


$$
A I = A
$$
$$
I A = A
$$


$$
\left(
    \begin{array}{c}
      1 & 2 \\
      3 & 4 \\
    \end{array}
\right)
\left(
    \begin{array}{cc}
      {1} & {0} \\
      {0} & {1} \\
    \end{array}
\right)
=
\left(
    \begin{array}{c}
      1 & 2 \\
      3 & 4 \\
    \end{array}
\right)
$$



## `逆行列`　（ inverse matrix ）
- *スカラーの逆数に対応*


$$
2 * 2^1 = 1
$$

$$
2　\frac{1}{2} = 1
$$


- **Aの逆行列**
    - *Aインバースと読む*
        - $I$ : 単位行列
        

$$

$$

$$
A^{-1}　A　=　I
$$


### <u>条件</u>
- **A が `正方行列` である事（最低限）**
    - 縦横size（行, 列）

# ベクトルで微分
### <u>file</u>
- `04_differential calculus`

$$
C=
\left(
    \begin{array}{c}
      3 \\
      4 \\
    \end{array}
\right)
$$


$$
x=
\left(
    \begin{array}{c}
      x_1 \\
      x_2 \\
    \end{array}
\right)
$$
- *上記の場合の$C^Tx$を計算する*


$$
C^Tx = \left(
    \begin{array}{c}
      3 & 4 \\
    \end{array}
\right)
\left(
    \begin{array}{c}
      x_1 \\
      x_2 \\
    \end{array}
\right)
=
3x_1 + 4x_2
$$

- **偏微分**

$$
\frac{\partial}{\partial a}( )
$$

- ディー（ラウンド）と呼ばれる
- a で偏微分する。　( ) は対象
    - a 以外を定数だと仮定して微分する
    - a 以外関係ないものは外に出して計算する
    - a に関係ないものは 0 に置き換えて計算


$$
\frac{\partial}{\partial x}(C^Tx)
=\left(
    \begin{array}{c}
    \frac{\partial}{\partial x_1}(3x_1 + 4x_2) \\
    \frac{\partial}{\partial x_2}(3x_1 + 4x_2) \\
    \end{array}
\right)
$$

- **$x_1$ の回答**
    - *$x_1$以外のものは外に出して計算*
        - <u>$x_1$に関係ないものは 0 に置き換える</u>


$$
\frac{\partial}{\partial x_1}(3x_1 + 4x_2)=\frac{\partial}{\partial x_1}(3x_1) + \frac{\partial}{\partial x_1}(4x_2)
$$


$$
\frac{\partial}{\partial x_1}(3x_1 + 4x_2)=3\frac{\partial}{\partial x_1}(x_1) + 4x_2\frac{\partial}{\partial x_1}(1)
$$


$$
\frac{\partial}{\partial x_1}(3x_1 + 4x_2)=(3*1) + (4x_2 * 0)
$$


$$
\frac{\partial}{\partial x_1}(3x_1 + 4x_2)=3
$$


- **$x_2$ の回答**
    - *$x_2$以外のものは外に出して計算*
        - <u>$x_2$に関係ないものは 0 に置き換える</u>


$$
\frac{\partial}{\partial x_2}(3x_1 + 4x_2)=\frac{\partial}{\partial x_2}(3x_1) + \frac{\partial}{\partial x_2}(4x_2)
$$


$$
\frac{\partial}{\partial x_2}(3x_1 + 4x_2)=3x_1\frac{\partial}{\partial x_2}(1) + 4\frac{\partial}{\partial x_2}(x_2)
$$


$$
\frac{\partial}{\partial x_2}(3x_1 + 4x_2)=(3x_1 * 0 * 1) + (4 * 1 * 1)
$$


$$
\frac{\partial}{\partial x_2}(3x_1 + 4x_2)=4
$$
