```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Graph ( scatter, plot, bar )

```python
df = pd.read_excel('./resources/Concentration.xlsx', index_col=0)

df.head()
```

```python
plt.scatter(df.index, df['C[mol/L]'])
plt.xlabel('time[s]', fontsize=16)
plt.ylabel('C[mol/L]', fontsize=16)
```

```python
plt.plot(df.index, df['C[mol/L]'])
```

```python
plt.bar(df.index, df['C[mol/L]'])
```

```python
x = np.arange(0, 15, 0.01)
y = np.exp(-0.1*x)
plt.bar(df.index, df['C[mol/L]'])
plt.plot(x, y, color='r')
```

## Graph ( subplot )

```python
x = np.arange(-10, 10, 0.01)
y = 1.0 / (1.0 + np.exp(-x))
```

```python
x2 = np.arange(-6, 6, 0.01)
y2 = np.sin(x2)
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.subplot(2, 1, 2)
plt.plot(x2, y2)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 4)
plt.plot(x2, y2)
```

## 3次元　plot

```python
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
```

```python
# キムサム用
a = np.arange(10)
a
```

```python
# キムサム : 累積和　：　
a.cumsum()
```

- **累積和**
    - *右側を足し合わせて次の解を算出する。それをどんどん左側にずらして行く*

```python
# 正規分布のキムサム
x = np.random.randn(100).cumsum()
y = np.random.randn(100).cumsum()
z = np.random.randn(100).cumsum()
```

```python
# 3Dには下記の手続きか必要
%matplotlib inline
%matplotlib notebook
fig = plt.figure()    # fig を描くような枠
ax = Axes3D(fig)      # 軸が　３Dに変更される
ax.plot(x, y, z)
plt.show()
```
