## 多項式近似・非線形近似

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
plt.style.use('ggplot')
np.random.seed(123)
```

### 多項式近似

```python
x = np.arange(20)
y = 2.0 * x
y_noise = np.random.randn(20)
y = y + y_noise
plt.scatter(x, y)
```

```python
param = np.polyfit(x, y, 1)
param
```

```python
plt.plot(x, param[0]*x + param[1])
plt.scatter(x, y)
```

### 非線形近似

```python
x2 = np.arange(15)
y2 = np.exp(0.4 * x2)
y2 = y2 + 10.0 * np.random.randn(15)
plt.scatter(x2, y2)
```

```python
def nonlinear_fit(x, a, b, c):
    return a * np.exp(b * x) + c
```

```python
popt, pcov = curve_fit(nonlinear_fit, x2, y2)
```

```python
popt
```

```python
plt.scatter(x2, y2)
plt.plot(x2, nonlinear_fit(x2, popt[0], popt[1], popt[2]))
```

## 数値積分

```python
from scipy import integrate
```

```python
x = np.arange(-100, 100, 0.01)
y = np.exp(-x ** 2)
plt.plot(x, y)
plt.xlim([-5, 5])
```

```python
integrate.simps(y, dx=0.01)
```

- **解析解**

$$\int_{-\infty}^{\infty}\exp(-ax^2)dx=\sqrt{\frac{\pi}{a}}$$

```python
np.sqrt(np.pi)
```

```python
# 積分の過程を知りたい場合
integrate.cumtrapz(y, dx=0.01)
```
