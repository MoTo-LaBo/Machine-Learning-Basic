## DataFrame 変更

```python
df_p = pd.read_excel('./resources/physical_measurement2.xlsx', index_col=0) # index_col=0 index 0番の 　id が index として使用される
df_p.head()
```

```python
# index を降順に並び替える
df = df_p.sort_index()
df.head()
```

```python
# NaN に対する対処 / NaN 値を削除
df.dropna()
```

```python
# NaN を置き換える
df.fillna(value=-1.0)
```

```python
# 直前の値を copy
df.fillna(method='ffill')
```

```python
# 平均値で　NaN を置き換える
df3 = df.fillna(value=df.mean())
df3
```

```python
# 重複値を削除
df4 = df3.drop_duplicates()
df4
```

```python
# 性別　: sex を数値化 / maleの場合は True = 1(真)　: False = -1(偽)
df4['sex2'] = df4['sex'].apply(lambda x : np.where(x == 'male', 1, -1))
```

```python
df4
```

```python
del df4['sex']
df4
```

## binning

```python
df_x = pd.read_csv('./resources/brownian_x.csv', delimiter=',', index_col=0)
df_y = pd.read_csv('./resources/brownian_y.csv', delimiter=',', index_col=0)
```

```python
df_d = (df_x ** 2 + df_y ** 2) ** 0.5 
```

```python
df_d.head()
```

```python
df_d.iloc[-1].hist(bins=25)
```

```python
bins = np.arange(0, 90, 10)
bins
```

```python
data = pd.cut(df_d.iloc[-1], bins)
data
```

```python
pd.value_counts(data)
```

```python
pd.value_counts(data, sort=False)
```

## Groupby

```python
plt.style.use('ggplot')
```

```python
df = pd.read_excel('./resources/multiple_data.xlsx')
df.head()
```

```python
groups = df.groupby("param")
for param, group in groups:
    plt.scatter(group['X'], group['Y'], label=param)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
df_1 = groups.get_group(1)
df_1
```

- **groupbyした後であれば**
    - *個別の param Number を　**get_group** で取得できる*
