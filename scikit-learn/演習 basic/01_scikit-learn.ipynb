## 入力変数と出力変数の切り分け

```python
df.head(3)
```

```python
# df.iloc[行, 列]
X = df.iloc[:, :-1]
X
```

```python
y = df.iloc[:, -1:]
y
```

## model の構築と検証

```python
# scikit-learn import : 重回帰分析のみ
from sklearn.linear_model import LinearRegression

# model 宣言
model = LinearRegression()

# model の学習
model.fit(X, y)
```

```python
# 検証 (決定係数の計算)
model.score(X, y)
```

## 訓練データと検証データの分割
- 訓練データ　:　train
- 検証データ　:　test

- `ダメなケース`
    - 受験勉強用に１０年分の過去問を購入
        - 10年すべて勉強(学力)
        - 10年分で実力test(検証)
        - *同じモノで test しても意味がない*
        


- `良いケース`
    - 前半の5年分で勉強（学習） <-　訓練data
    - 後半の5年分で実力test(検証)　<-　検証data
    


- <u>Scikit-learn でその機能が実装されている</u>

```python
from sklearn.model_selection import train_test_split
```

```python
# 訓練データと検証データの分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)  # random seed(1) 再現性の固定
```

```python
# model の学習 : paramertor の調整
model.fit(X_train, y_train)
```

```python
# 検証 <- 検証データ
model.score(X_test, y_test)
```

```python
# 検証　<- 訓練データ
model.score(X_train, y_train)
```

- *基本的には訓練データの方が精度が良くて、検証データでは少し精度が落ちる*
    - **数値が近ければ精度は高い。あとはどこまで数値を求めるか**
