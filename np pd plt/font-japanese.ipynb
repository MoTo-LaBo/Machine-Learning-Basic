````python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib          # matplolib 日本語対応
%matplotlib inline
````

````python
dalta = 0.2
x = y = np.arange(-10.0, 10.0, dalta)
xx, yy = np.meshgrid(x, y)
z = xx**2 - yy**2

# 描画サイズ (単位はインチ)
fig = plt.figure(figsize=(6, 6))

# 三次元グラフ描画 subplot を生成
ax = fig.gca(projection='3d')

# 等高線の表示
ax.contour3D(xx, yy, z, levels=40, cmap='Blues')

# x, y, z の描画範囲の設定
ax.set_xlim(xx.min()-1, xx.max()+1)     # x の領域より少しマージを持たせている
ax.set_ylim(yy.min()-1, yy.max()+1)     # y の領域よりも少しマージを持たせている
ax.set_zlim(-abs(z).max(), abs(z).max())  # z の絶対値の最大値としその反対を最小値とする
ax.set_title('3次元グラフ', fontsize=14)
plt.show()
````
